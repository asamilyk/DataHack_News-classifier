```python
import pandas as pd
from catboost import CatBoostClassifier, cv
from sklearn.model_selection import train_test_split
```

```python
X_raw = pd.read_csv('../proba.csv')
Copy = X_raw.copy()
display(X_raw)
```

```python
import nltk
from pymystem3 import Mystem
import pymystem3

mystem_analyzer = Mystem()
```

```python
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
```

```python
X_raw['Title'] = X_raw['Title'].apply(lambda x: ''.join(mystem_analyzer.lemmatize(x)))
```

```python
for i in range(0, 9):  
    X_raw[str(i)] = X_raw[str(i)].apply(lambda x: ''.join(mystem_analyzer.lemmatize(str(x))).strip())
```

Вместо Null в категории - ""

```python
y = X_raw['Category']
X = X_raw.drop(['Category', 'URL', 'Article Text', 'Description', 'Unnamed: 0'], axis=1)
# Загрузка данных
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)

X
```

```python
# Явное указание категориальных признаков
cat_features = list(X_train.columns)

# Создание и обучение модели с указанием категориальных признаков
model = CatBoostClassifier(iterations=100, learning_rate=0.1, depth=7)
model.fit(X_train, y_train, cat_features=cat_features)

# Оценка производительности модели
accuracy = model.score(X_test, y_test)
print(f"Accuracy: {accuracy}")
```

```python
model.save_model("model.pkl")
```
